```python
from igraph import *
import numpy as np
```

```python
g1 = Graph.Erdos_Renyi(10,0.3,directed=False)
g1.vcount()
```

```python
g2 = Graph.Erdos_Renyi(12,0.5,directed=False)
g2.vcount()
```

```python
g = g1+g2
g.vcount()
```

```python
g.ecount()
```

```python
igraph.plot(g, bbox = (400,400), layout = g.layout('kk'))
```

```python
g.get_edgelist()
```

```python
g.add_edge(1,21)
```

```python
g.get_edgelist()
```

```python
plot(g, layout = g.layout_kamada_kawai())
```

```python
visual_style = {}
visual_style["vertex_size"] = 20
visual_style["layout"] = g.layout_kamada_kawai()
visual_style["bbox"] = (300, 300)
visual_style["margin"] = 20
```

```python
plot(g, **visual_style)
```
